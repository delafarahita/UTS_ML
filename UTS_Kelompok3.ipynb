Langkah 2 - Load Data

```python
import os
import numpy as np
from skimage.feature import hog
from PIL import Image, UnidentifiedImageError  # Menggunakan UnidentifiedImageError dari PIL

def load_images(data_dir, image_size=(128, 128), min_images_per_person=10, max_images_per_person=100, max_person=50):
    # List untuk fitur HOG dan label
    X_hog_features = []
    labels = []
    class_names = []
    person_count = 0  # Menghitung jumlah orang yang diproses

    # Iterasi semua folder dan file dalam dataset
    for root, dirs, files in os.walk(data_dir):
        # Memastikan folder memiliki cukup gambar dan jumlah orang tidak melebihi batas
        if len(files) >= min_images_per_person and person_count < max_person:
            class_name = os.path.basename(root)

            # Tambahkan kelas baru jika belum ada
            if class_name not in class_names:
                class_names.append(class_name)
                person_count += 1
            
            label = class_names.index(class_name)
            print(f"Loading images from {root}...")  # Debugging log

            # Batas jumlah gambar yang diproses per orang
            image_count = 0
            for img_name in files:
                if image_count >= max_images_per_person:
                    break  # Hentikan jika mencapai batas gambar per orang

                img_path = os.path.join(root, img_name)
                
                if os.path.isfile(img_path):
                    try:
                        with Image.open(img_path) as img:
                            img = img.convert('L').resize(image_size)  # Convert to grayscale and resize
                            img_array = np.array(img) / 255.0  # Normalize pixel values

                            # Ekstraksi fitur HOG
                            hog_features = hog(img_array, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=False)
                            X_hog_features.append(hog_features)

                            labels.append(label)
                            image_count += 1  # Tambah jumlah gambar yang valid
                            
                    except (UnidentifiedImageError, OSError) as e:
                        print(f"Skipping {img_name}: {e}")

    return np.array(X_hog_features), np.array(labels), class_names

# Panggil fungsi load_images dengan path dataset yang sudah diunduh
path = kagglehub.dataset_download("ntl0601/casia-webface")  # Ganti dengan path dataset Anda
X_hog_features, labels, class_names = load_images(path, min_images_per_person=5, max_images_per_person=100, max_person=10)

# Konversi ke array NumPy jika perlu
print("HOG Features Shape:", X_hog_features.shape)
print("Labels Shape:", labels.shape)
print("Labels:", labels)
```

Langkah 3 - Split Data

Langkah 2 - Fitting Model

Evaluasi Model untuk Fitur HOG

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Asumsi bahwa X_hog_features dan labels telah dihasilkan sebelumnya
# Standarisasi data HOG
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_hog_features)

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(labels)

# Bagi data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Hitung jumlah kelas unik
num_classes = len(np.unique(y_train))

# Definisikan model ANN
input_shape = (X_train.shape[1],)  # Input shape sesuai dengan panjang fitur HOG
ann = Sequential()
ann.add(Dense(units=64, activation='relu', input_shape=input_shape))
ann.add(Dense(units=32, activation='relu'))
ann.add(Dense(num_classes, activation='softmax'))  # Gunakan num_classes sesuai dengan jumlah kelas unik

# Compile model
ann.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training model
ann.fit(X_train, y_train, batch_size=32, epochs=100, validation_data=(X_test, y_test))
```

```python
# Evaluasi model ANN menggunakan data test HOG
loss, accuracy = ann.evaluate(X_test, y_test)
print(f"Test Loss (HOG): {loss}")
print(f"Test Accuracy (HOG): {accuracy * 100:.0f}%")
```
